```python
def calc(feature):
    mean = feature.mean()
    median = feature.median()
    mode = feature.mode()
    var = feature.var()
    std = feature.std()
    min = feature.min()
    max = feature.max()
    Q1, Q3 = feature.quantile(0.25), feature.quantile(0.75)
    IQR = Q3 - Q1
    # print(mode)
    return mean, median, mode, var, std, min, max, Q1, Q3, IQR
```

```python

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Tạo một Figure (khung tranh) chứa lưới 2 hàng x 2 cột
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten() # Dàn phẳng lưới để dễ dùng vòng lặp

# Lặp qua từng đặc trưng và vẽ Boxplot
for i, feature in enumerate(features):
    # Dùng x='species' để chia nhóm, y=feature để lấy giá trị
    sns.boxplot(
        data=df, 
        x='species', 
        y=feature, 
        hue='species',      # Tô màu theo loài
        palette='Set2',     # Chọn dải màu dịu mắt
        legend=False,       # Tắt chú thích phụ vì trục x đã hiện tên loài
        ax=axes[i]
    )
    
    # Trang trí biểu đồ
    axes[i].set_title(f'So sánh {feature} theo loài', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('Loài hoa (Species)')
    axes[i].set_ylabel('Kích thước (cm)')

# Tự động căn chỉnh khoảng cách giữa các biểu đồ
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_prevalence_impact():
    # Tạo mảng P(B) từ 0.001 đến 0.2
    P_B_array = np.linspace(0.001, 0.2, 500)
    P_not_B_array = 1 - P_B_array
    
    P_pos_given_B = 0.99
    P_pos_given_not_B = 0.05
    
    # Tính P(B|+) cho từng giá trị của P(B)
    P_pos = (P_pos_given_B * P_B_array) + (P_pos_given_not_B * P_not_B_array)
    P_B_given_pos_array = (P_pos_given_B * P_B_array) / P_pos
    
    # Vẽ đồ thị
    plt.figure(figsize=(10, 6))
    plt.plot(P_B_array, P_B_given_pos_array, color='blue', linewidth=2)
    plt.title("Xác suất thực sự mắc bệnh P(B|+) theo Tỉ lệ mắc bệnh P(B)")
    plt.xlabel("Tỉ lệ mắc bệnh P(B)")
    plt.ylabel("Xác suất mắc bệnh khi test dương tính P(B|+)")
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Đánh dấu điểm P(B) = 0.01 như bài toán gốc
    plt.plot(0.01, 0.1667, 'ro')
    plt.annotate('P(B)=0.01, P(B|+)=16.67%', xy=(0.01, 0.1667), xytext=(0.03, 0.1),
                 arrowprops=dict(facecolor='red', shrink=0.05))
    plt.show()

# Chạy hàm để hiển thị đồ thị
# plot_prevalence_impact()
```

```python
def naive_bayes_spam_filter(email_words):
    # Xác suất tiên nghiệm (Prior probabilities)
    P_Spam = 0.5
    P_Ham = 0.5
    
    # Xác suất có điều kiện (Likelihoods)
    likelihoods = {
        "khuyến mãi": {"Spam": 0.8, "Ham": 0.1},
        "trúng": {"Spam": 0.9, "Ham": 0.05},
        "họp": {"Spam": 0.05, "Ham": 0.8}
    }
    
    # Tính tử số cho Spam và Ham
    score_spam = P_Spam
    score_ham = P_Ham
    
    for word in email_words:
        if word in likelihoods:
            score_spam *= likelihoods[word]["Spam"]
            score_ham *= likelihoods[word]["Ham"]
            
    # Chuẩn hóa để tổng xác suất bằng 1 (Tính P(W) làm mẫu số)
    P_W = score_spam + score_ham
    
    P_Spam_given_W = score_spam / P_W
    P_Ham_given_W = score_ham / P_W
    
    return P_Spam_given_W

# Khảo sát 2 email ví dụ
email_1 = ["khuyến mãi", "trúng"] 
email_2 = ["họp", "khuyến mãi"]

print(f"Email 1 {' '.join(email_1)}:")
print(f"  -> Xác suất Spam: {naive_bayes_spam_filter(email_1) * 100:.2f}%")

print(f"\nEmail 2 {' '.join(email_2)}:")
print(f"  -> Xác suất Spam: {naive_bayes_spam_filter(email_2) * 100:.2f}%")
```
